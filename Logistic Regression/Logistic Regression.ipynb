# 1. LOGISTIC REGRESSION MODEL

```r
# Tao tap train/test

n = nrow(data)
trainIndex = sample(1:n, size = round(0.7*n), replace = FALSE)
train = data[trainIndex,]
test = data[-trainIndex,]
print('Selected train row:')
print(nrow(train))
print('Selected test row:')
print(nrow(test))
```

```r
# Build model
mylogit = glm(Class ~ . , data = train, family = 'binomial')
print(summary(mylogit))
```

Nhận xét:
Có khá nhiều biến cho kết quả rất tốt (***)

```r
# Sử dụng model để apply cho test data

pred = predict(mylogit, newdata = test, type = 'response')

threshold = 0.5
pred_value = ifelse (pred > threshold, 1, 0)

# In 1 vài kết quả để xem
print('Test data Class vs predict (10 rows:)')
result = data.frame(data_class = test$Class[30:40], pred_value[30:40])
print(result)
```

```r
# danh gia model
confusionMatrix(as.factor(pred_value),as.factor(test$Class))
```

Nhận xét: Accuracy cao 94%, p-value thấp 0.05--> Cho thấy model khá hiệu quả.
